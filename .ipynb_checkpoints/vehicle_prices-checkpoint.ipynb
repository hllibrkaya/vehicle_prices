```python
def show_duplicates(dataframe):
    """
    Identify and display duplicate rows in a pandas DataFrame.

    Parameters:
    dataframe (pandas.DataFrame): The DataFrame to search for duplicates.

    Returns:
    None

    Notes:
    This function calculates the number of duplicated rows in the provided DataFrame 
    and prints information about them. If duplicates are found, it displays those 
    rows as well.

    """

    sum = dataframe.duplicated().sum()
    print(f"There is {sum} duplicated rows")
    if sum > 0:
        print(dataframe.loc[dataframe.duplicated(keep=False), :])
```

```python
# Note that this is not scientific and some brands have some higher class models, but it is explaining car's general target class
def brand_to_category(brand):
    lux = ["Lexus", "Volvo", "BMW", "Jaguar", "Land", "Mercedes-Benz", "Audi", "Jeep", "MG"]
    middle = ["Isuzu", "Kia", "Toyota", "Force", "Mitsubishi", "Skoda", "Honda", "Ford",
              "Volkswagen", "Nissan"]
    economy = ["Chevrolet", "Maruti", "Hyundai", "Renault", "Maruti", "Tata", "Mahindra",
               "Datsun", "Fiat", "Ambassador", "Peugeot", "Ashok", "Opel", "Daewoo"]
    if brand in lux:
        return "Lux"
    elif brand in middle:
        return "Middle"
    elif brand in economy:
        return "Economy"
    else:
        return "Other"
```

```python
data["brand_category"] = data["brand"].apply(brand_to_category)
```

*repressing outliers seems inevitable, after that i might think to use log transformation for skewness*

```python
def supress_outliers(dataframe, column, q1=0.25, q3=0.75):
    quantile1 = dataframe[column].quantile(q1)
    quantile3 = dataframe[column].quantile(q3)
    iqr = quantile3 - quantile1
    upper_bound = quantile3 + 1.5 * iqr
    lower_bound = quantile1 - 1.5 * iqr
    dataframe.loc[dataframe[column] > upper_bound, column] = upper_bound
    dataframe.loc[dataframe[column] < lower_bound, column] = lower_bound
```

*it seems km column has outliers too, so lets check outliers fastly with a function*

```python
def outlier_plots(dataframe, column):
    quantile1 = dataframe[column].quantile(0.10)
    quantile3 = dataframe[column].quantile(0.80)
    iqr = quantile3 - quantile1
    upper_bound = quantile3 + 1.5 * iqr
    lower_bound = quantile1 - 1.5 * iqr
    if dataframe[(dataframe[column] > upper_bound) | (dataframe[column] < lower_bound)].any(
            axis=None):  #if there are any outlier value
        sns.boxplot(data=dataframe, x=column)
        plt.title(f"Outliers in {column}")
        plt.ticklabel_format(style='plain', axis="x")
        plt.show()
```

```python
def missing_values_table(dataframe):
    """
    Calculate and display a summary of missing values in the given DataFrame.

    Parameters:
    -----------
    dataframe : pandas DataFrame
        DataFrame containing the data.

    Returns:
    --------
    None
        Displays a table summarizing the missing values.

    Example:
    --------
    import pandas as pd
    import numpy as np

    # Create a sample DataFrame with missing values
    data = {
        'A': [1, np.nan, 3, 4, 5],
        'B': [6, 7, np.nan, 9, 10],
        'C': [11, 12, 13, 14, np.nan]
    }
    df = pd.DataFrame(data)

    # Display summary of missing values
    missing_values_table(df)
    """
    na_columns = [col for col in dataframe.columns if dataframe[col].isnull().sum() > 0]

    total = dataframe.shape[0]

    n_miss = dataframe[na_columns].isnull().sum().sort_values(ascending=False)

    ratio = (dataframe[na_columns].isnull().sum() / total * 100).sort_values(ascending=False)
    entries = total - n_miss

    missing_df = pd.concat([entries, n_miss, np.round(ratio, 2)], axis=1, keys=["entries", 'num_miss', 'miss_ratio'])

    missing_df.insert(0, "total", total)

    print(missing_df, end="\n")
```

# Model Selection & Automated Hyper-Parameter Optimization

```python
lr_params = {"fit_intercept": [True, False]}
pl_params = {"polynomialfeatures__degree": list(range(1, 5, 1))}
poly_pipe = make_pipeline(PolynomialFeatures(), LinearRegression())
alpha_params = {"alpha": np.arange(0.1, 1000.0, 0.5)}
elastic_params = {"alpha": np.arange(0.1, 1000.0, 0.5), "l1_ratio": np.arange(0.1, 1.0, 0.1)}
ps_params = {'alpha': np.arange(0.1, 1.1, 0.1), 'max_iter': list(range(50, 500, 50))}
rf_params = {"n_estimators": list(range(50, 500, 50)), "max_depth": [None, 10, 20], "min_samples_split": [2, 5, 10],
             "min_samples_leaf": [1, 2, 4]}
dt_params = {"max_depth": [None, 10, 20], "min_samples_split": [2, 5, 10], "min_samples_leaf": [1, 2, 4]}
gb_params = {"n_estimators": list(range(100, 500, 50)), "learning_rate": [0.05, 0.1, 0.2], "max_depth": [3, 5, 7],
             "max_features": [0.5, 0.75, 1.0]}
ada_params = {"n_estimators": [50, 100, 200], "learning_rate": [0.05, 0.1, 0.2],
              "estimator": [LinearRegression(), Ridge(), Lasso()]}
bagging_params = {"n_estimators": [10, 20, 50], "max_samples": [0.5, 1.0], "max_features": [0.5, 0.75, 1.0]}
svr_params = {"kernel": ["linear", "poly", "rbf"], "C": np.arange(200, 600.1, 20), "gamma": ["scale", "auto"]}
lgbm_params = {"n_estimators": list(range(100, 500, 50)), "learning_rate": [0.05, 0.1, 0.2], "max_depth": [3, 5, 7],
               "num_leaves": [31, 50, 100], "objective": ["regression", "regression_l1"], "verbose": [-1]}

xg_params = {'n_estimators': list(range(100, 500, 50)), "learning_rate": [0.05, 0.1, 0.2], "max_depth": [3, 5, 7],
             "objective": ["reg:squarederror"], "subsample": np.arange(0.2, 0.9, 0.2), "verbosity": [0]}

cat_params = {"iterations": list(range(100, 500, 50)), "learning_rate": [0.05, 0.1, 0.2], 'depth': [3, 5, 7],
              "verbose": [0]}
```

```python
models = [("Linear Regression", LinearRegression(), lr_params),
          ("Polynomial Regression", poly_pipe, pl_params),
          ("Ridge", Ridge(), alpha_params),
          ("Lasso", Lasso(), alpha_params),
          ("EleasticNet", ElasticNet(), elastic_params),
          ("Poisson", PoissonRegressor(), ps_params),
          ("Decision Tree", DecisionTreeRegressor(), dt_params),
          ("Random Forest", RandomForestRegressor(), rf_params),
          ("Gradient Boosting", GradientBoostingRegressor(), gb_params),
          ("Bagging", BaggingRegressor(), bagging_params),
          ("AdaBoost", AdaBoostRegressor(), ada_params),
          #("SVR", SVR(), svr_params), #it takes too long after log. transformation
          ('LightGBM', LGBMRegressor(verbose=-1), lgbm_params),
          ('XGBoost', XGBRegressor(use_label_encoder=False, objective="reg:squarederror", verbosity=0), xg_params),
          ("CatBoost", CatBoostRegressor(verbose=0), cat_params)
          ]
```

```python
def hyperparameter_optimization(X, y, models, cv=5, scoring="r2"):
    """
    Perform hyperparameter optimization for machine learning models using cross-validation.

    Parameters:
    -----------
    X : array-like or sparse matrix, shape (n_samples, n_features)
        Training data.

    y : array-like, shape (n_samples,) or (n_samples, n_targets)
        Target values.

    models : list of tuples
        List containing tuples of model name, model object, and corresponding hyperparameter grid.

    cv : int, default=5
        Number of folds for cross-validation.

    scoring : str, default="r2"
        Scoring metric to optimize during hyperparameter tuning.

    Returns:
    --------
    best_models : dict
        Dictionary containing the best models found after hyperparameter optimization.
        Keys are model names and values are dictionaries with "Model" key for the optimized model object
        and "R2 Score" key for the mean R2 score obtained through cross-validation.

    Example:
    --------
    import numpy as np
    from sklearn.datasets import load_diabetes
    from sklearn.ensemble import RandomForestRegressor
    from sklearn.linear_model import Lasso
    from sklearn.model_selection import GridSearchCV

    # Load dataset
    diabetes = load_diabetes()
    X = diabetes.data
    y = diabetes.target

    # Define models with hyperparameter grids
    models = [
        ("Random Forest", RandomForestRegressor(), {'n_estimators': [10, 50, 100], 'max_depth': [None, 5, 10]}),
        ("Lasso Regression", Lasso(), {'alpha': np.logspace(-4, 4, 10)})
    ]

    # Perform hyperparameter optimization
    best_models = hyperparameter_optimization(X, y, models, cv=5, scoring="r2")
    """
    print("Hyperparameter Optimization....")
    best_models = {}
    for name, model, params in models:
        print(f"########## {name} ##########")
        cv_results = cross_val_score(model, X, y, cv=cv, scoring=scoring, n_jobs=-1)
        print(f"{scoring} (Before): {cv_results.mean()}")

        gs_best = GridSearchCV(model, params, cv=cv, n_jobs=-1, verbose=False).fit(X, y)
        final_model = model.set_params(**gs_best.best_params_)

        cv_results = cross_val_score(final_model, X, y, cv=cv, scoring=scoring, n_jobs=-1)
        print(f"{scoring} (After): {cv_results.mean()}")
        print(f"{name} best params: {gs_best.best_params_}", end="\n\n")
        best_models[name] = {"Model": final_model, "R2 Score": cv_results.mean()}
    return best_models
```

```python
best_models = hyperparameter_optimization(X_train, y_train, models)
```

```python
df = pd.DataFrame(best_models).transpose()
sns.barplot(data=df, x='R2 Score', y=df.index, palette="viridis")
plt.title('R2 Scores After Hyperparameter Optimization')
plt.xlabel('Model')
plt.ylabel('R2 Score')
```
